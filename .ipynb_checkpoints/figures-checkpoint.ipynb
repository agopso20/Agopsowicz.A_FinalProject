```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as md
# reading and opening temperature data from spreadsheet
filepath1='HPIP_Yakutsk_full.xls'
nas=(-999.0,-999)
yktsk_temp_df=pd.read_excel(filepath1,
                 sheet_name='Jan soil & surface  temp',
                 header=10,usecols=['YEAR','0.20 m','3.20 m','ds 570 surface temp'],
                 skiprows=range(11,48),skipfooter=6,na_values=nas,na_filter=True)
# converted years column to datetimes (kathy helped with this and the NaNs)
yktsk_temp_df['YEAR']=pd.to_datetime(yktsk_temp_df['YEAR'], format='%Y').dt.date 
# used loc to fix the -999 nans that were read as integers instead of floats
yktsk_temp_df.loc[yktsk_temp_df['ds 570 surface temp'] == -999, 'ds 570 surface temp']=np.nan
display(yktsk_temp_df)

# plotting temp data
fig,ax=plt.subplots(figsize=(12,6))
x_temp=yktsk_temp_df['YEAR'].values
y_temp=yktsk_temp_df['ds 570 surface temp'].values

plt.plot(x_temp,y_temp,marker='.',color='lightslategray',label='ds 570 Surface Temperature (deg C)')
# manipulated datetime frequency with datetime formatter to fit all years legibly on my x-axis
ax.xaxis.set_major_locator(md.YearLocator(2))
ax.xaxis.set_major_formatter(md.DateFormatter('%Y'))
plt.xticks(rotation=45, ha='right')

ax.set_ylabel('ds 570 Surface Temperature (deg C)')
ax.set_xlabel('Years')
ax.grid()
ax.set_title('Yearly Yakutsk Soil Surface Temperature 1951-2003')
```

```python
#opening and reading coastal retreat data for site 3, started with this site before deciding to analyze all 3
#used pd.read_table to read tab file
filepath2='Lena_Delta_coastal_retreat_1951-2018.tab'
lena_retr_df=pd.read_table(filepath2, header=15, skiprows=range(16,82), 
                           usecols=['Event', 'CRR [m/a]', 'Date/Time'])
#converting to datetimes
lena_retr_df['Date/Time']=pd.to_datetime(lena_retr_df['Date/Time'], format='%Y').dt.date
display(lena_retr_df)

#plotting coastal retreat on its own to check that I opened dataframe correctly
fig,ax=plt.subplots(figsize=(12,6))
x_ts3=lena_retr_df['Date/Time'].values
y_ts3=lena_retr_df['CRR [m/a]'].values

ax.plot(x_ts3,y_ts3,marker='.')
# manipulated datetime frequency with datetime formatter to fit all years legibly on my x-axis
ax.xaxis.set_major_locator(md.YearLocator(2))
ax.xaxis.set_major_formatter(md.DateFormatter('%Y'))
plt.xticks(rotation=45, ha='right')

ax.grid()
ax.set_ylabel('Coastal Retreat Rate [m/a]')
ax.set_xlabel('Years')
ax.set_title('Yearly Lena Delta Coastal Retreate Rate 1982-2018') 
```

```python
from datetime import datetime
# plotting site 3 against temperature data
fig,ax=plt.subplots(figsize=(14,6))
line1_ax=ax.plot(x_ts3,y_ts3,marker='.',color='tomato',label='Bykovsky Site 3')

# set axis limits to show overlapping data
# manipulated datetime frequency with datetime formatter to fit all years legibly on my x-axis
ax.set_xlim(datetime(1982,1,1),datetime(2003,1,1))
ax.set_ylim(2,8)
ax.xaxis.set_major_locator(md.YearLocator(1))
ax.xaxis.set_major_formatter(md.DateFormatter('%Y'))
plt.xticks(rotation=45, ha='right')

ax.grid()
ax.set_ylabel('Coastal Retreat Rate [m/a]')
ax.set_xlabel('Years')

# twinned y-axis to plot temperature data on separate scale
twin=ax.twinx()
line2_ax=twin.plot(x_temp,y_temp,marker='.',color='lightslategray',label='ds 570 Surface Temperature (deg C)')
twin.set_ylabel('ds 570 Surface Temperature (deg C)')
ax.set_title('Overlapping Yearly Temperature and Coastal Retreat Data')

# joined lines of each data plot and used for statement to join my labels in one legend
lines_ax=line1_ax+line2_ax
labs_ax = [l.get_label() for l in lines_ax]
ax.legend(lines_ax,labs_ax,loc='upper left')
plt.show()
```
